```python
# EXERCISE 1
import matplotlib.pyplot as plt
City_A =[500000, 550000, 600000, 650000, 700000, 750000, 800000] 
City_B =[800000, 850000, 900000, 950000, 1000000, 1050000, 1100000] 
City_C =[1000000, 1050000, 1100000, 1150000, 1200000, 1250000, 1300000] 
City_D =[1200000, 1250000, 1300000, 1350000, 1400000, 1450000, 1500000]
X_axis=[2010,2011,2012,2013,2014,2015,2016]
plt.plot(X_axis,City_A,label='CITY A')
plt.plot(X_axis,City_B,label='CITY B')
plt.plot(X_axis,City_C,label='CITY C')
plt.plot(X_axis,City_D,label='CITY D')
plt.xlabel=('year')
plt.ylabel=('population')
plt.title=('Population Growth')
plt.legend()
plt.grid()
plt.show()
```
